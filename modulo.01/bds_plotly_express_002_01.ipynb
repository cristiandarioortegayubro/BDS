 # **<font color="DeepPink">Gr치fico de barras</font>**

<p align="justify">
Un gr치fico de barras es un tipo de visualizaci칩n que utiliza barras rectangulares para representar datos. En un gr치fico de barras, la longitud de cada barra es proporcional al valor que representa. Estos gr치ficos son 칰tiles para comparar cantidades entre diferentes categor칤as o grupos.
<br><br>
Aqu칤 hay algunos aspectos clave de los gr치ficos de barras:
<br><br>

1. **Ejes:** Los gr치ficos de barras generalmente tienen un eje horizontal (eje x) que representa las categor칤as o grupos de datos, y un eje vertical (eje y) que representa la magnitud o la cantidad de los datos.

2. **Barras:** Cada barra en el gr치fico representa una categor칤a o grupo de datos, y la altura o longitud de la barra corresponde al valor que representa. Las barras pueden ser horizontales o verticales, dependiendo de la orientaci칩n del gr치fico.

3. **Comparaci칩n:** Los gr치ficos de barras son especialmente 칰tiles para comparar valores entre diferentes categor칤as de datos. Por ejemplo, pueden utilizarse para comparar ventas por producto, ingresos por regi칩n, o frecuencia de eventos en diferentes per칤odos de tiempo.

4. **Tipos:** Hay varios tipos de gr치ficos de barras, incluyendo gr치ficos de barras simples, gr치ficos de barras apiladas (donde las barras se apilan una encima de otra), y gr치ficos de barras agrupadas (donde las barras se agrupan en categor칤as).

<br>
<p align="justifi">
Los gr치ficos de barras son una herramienta com칰n en la visualizaci칩n de datos y se utilizan ampliamente en diversas 치reas, como negocios, investigaci칩n, ciencia, y m치s.

```python
import pandas as pd
```

```python
# Datos de muestra
df = pd.DataFrame(dict(
    grupo = ["A", "B", "C", "D", "E"],
    valor = [14, 12, 8, 10, 16],
    error = [1.4, 2.4, 0.7, 1, 2],
    text =  ['G1', 'G2', 'G3', 'G4', 'G5']))
```

```python
df
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             template = "gridon")

fig.show()
```

```python
fig = px.bar(df,
             x = 'valor',
             y = 'grupo',
             template = "gridon")

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             error_y = 'error',
             template = "gridon")

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             title = "T칤tulo",
             template = "gridon",
             labels = {'grupo': 'Etiqueta de los grupos',
                       'valor': 'Etiqueta de los valores'})

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             color = 'grupo',
             template = "gridon")

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             color = 'grupo',
             template = "gridon",
             color_discrete_map = {'A': '#7FD4C1',
                                   'B': '#30BFDD',
                                   'C': '#8690FF',
                                   'D': '#ACD0F4',
                                   'E': '#F7C0BB'})

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             template = "gridon")

fig.update_traces(marker_color = 'green',
                  marker_line_color = 'black',
                  marker_line_width = 2,
                  opacity = 1)

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             color = 'valor',
             template = "gridon")

fig.show()
```

```python
fig = px.bar(df,
             x = 'grupo',
             y = 'valor',
             color = 'valor',
             color_continuous_scale = 'bluered',
             template = "gridon")

fig.show()
```

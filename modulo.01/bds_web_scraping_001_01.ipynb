# **<font color="DeepPink">Creando un DataFrame</font>**


<p align="justify">  Utilizamos el m茅todo <code>read_html()</code> de <code>Pandas</code>.
</p>

```python
lista = pd.read_html(data)
```

<p align="justify">  Vemos que el objeto al cual se le asigno la lectura del html, es una lista, y lo verificamos con <code>type()</code>. En este caso, esta lista tiene un solo elemento y ese elemento es un <code>DataFrame</code>, es decir, el m茅todo <code>read_html()</code> genera listas de $n$ elementos, al consultar o llamar a los elementos, se visualizan los <code>DataFrames</code>.

```python
type(lista)
```

```python
len(lista)
```

```python
df = lista[0]
df
```

<p align="justify">  Ahora colocamos los nombres de las columnas, teniendo en cuenta la etiqueta $1$ del indice. Pero al hacer eso, vemos que a las columnas se le asigna un nombre. Por ese motivo, se establece con la propiedad <code>columns.name</code> que las columnas no tengan nombre. Por otro lado, con el m茅todo <code>display()</code> se puede visualizar el <code>DataFrame</code> cuando queremos verlo entre operaciones que estamos desarrollando en nuestro script, como por ejemplo, la celda a continuaci贸n:
</p>

```python
df.columns = df.loc[1]
display(df)
df.columns.name = None
df
```

<p align="justify">  Ahora filtramos por los indices que queremos ver, que son los correspondientes a los datos mensuales, la recaudaci贸n de Enero a Diciembre. Esto directamente lo hacemos con la selecci贸n de indices, el m茅todo <code>loc[]</code>. La idea de este paso, es poder realizar a continuaci贸n, la eliminaci贸n de datos nulos.
</p>

```python
df = df.loc[2:13]
df
```

<p align="justify">  Ahora eliminamos las columnas que tienen todos sus datos nulos, ya que puede pasar que en el ejercicio fiscal en marcha, solo esten cargados algunos meses de recaudaci贸n, es decir, los meses anteriores al mes de la fecha que estamos consultando. Esos datos se ir谩n completando, a medida que pase el tiempo y lleguemos a fin de a帽o.
</p>

```python
df = df.dropna(axis=1, how="all")
df
```

<p align="justify">  Ahora seteamos el indice con la columna <code>Meses</code>, para dejar todo el <code>DataFrame</code> con valores "num茅ricos" pero vamos a verificar que tipo de datos tiene para ajustarlo al tipo de datos que corresponde, es decir, que nuestros datos sean num茅ricos.
</p>

```python
df.set_index("Meses", inplace=True)
df
```

<p align="justify">  Los datos de nuestro <code>DataFrame</code> no son numericos.
</p>

```python
df.info()
```

```python
df = df.astype("float")
```

<p align="justify">  Ahora si son num茅ricos decimales.
</p>

```python
df.info()
```

<p align="justify">  Vamos a crear un diccionario para colocar los nombres de los meses y su equivalente num茅rico. La idea es mapear en el <code>DataFrame</code> y convertir los nombres de los meses en n煤meros. De esa forma, al graficar, el eje $X$ va a estar m谩s ordenado.
</p>

```python
Meses = {"Enero":1, "Febrero":2, "Marzo":3, "Abril":4, "Mayo":5, "Junio":6,
         "Julio":7, "Agosto":8, "Septiembre":9, "Octubre":10, "Noviembre":11, "Diciembre":12}
```

Aplicamos el metodo <code>map()</code> al indice de nuestro <code>DataFrame</code>, y supuestamente con esto, ya estar铆a limpio todo nuestro <code>DataFrame</code>, pero lo que queremos hacer a continuaci贸n es reordenarlo solo a los efectos de graficar, ya que la distribuci贸n de los datos se encuentra a lo ancho, y <code>Plotly</code> utiliza <code>DataFrames</code> con formato a lo largo.

```python
df.index = df.index.map(Meses)
```

```python
df
```

```python
lista = []
```

```python
for i, j in enumerate(df.columns):
  datos = pd.DataFrame(df[j])
  datos = datos.rename(columns={j:"iibb"})
  datos["A帽o"] = 2000+i
  datos = datos[["A帽o","iibb"]]
  lista.append(datos)
```

```python
iibb = pd.concat(lista)
iibb.index.name = "Mes"
iibb.reset_index(inplace=True)
```

```python
iibb
```

# **<font color="DeepPink">Gr谩fico</font>**


```python
px.line(iibb.query("A帽o > 2002 & A帽o < 2005"),
        x="Mes",
        y="iibb",
        color="A帽o",
        template="gridon",
        title="Ingresos Brutos",
        markers=True,
        labels={"iibb":"Recaudaci贸n"})
```

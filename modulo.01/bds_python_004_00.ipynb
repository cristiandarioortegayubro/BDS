## **<font color="DeepPink">Declarando funciones</font>**

<p align="justify">
Las funciones se declaran con la palabra clave <code>def</code>. Una funci贸n contiene un bloque de c贸digo con un uso opcional de la palabra clave <code>return</code>.<br><br> La sintaxis quedar铆a as铆:


````python
act = 1000
pas = 600
pneto = 400
````

````python
def ecuacion_patrimonial(act, pas, pneto):
  print(f"El activo es de {act}, el pasivo de {pas} y el patrimonio neto de {pneto}")
  if act == (pas + pneto):
    print(f"El activo es igual al pasivo mas el patrimonio neto")
  else:
    print(f"No se cumple con la ecuaci贸n patrimonial")
````

````python
ecuacion_patrimonial(act=1200, pas=500, pneto=600)
````

<p align="justify">
Dado que las funciones de Python son objetos, se pueden expresar f谩cilmente muchas construcciones que son dif铆ciles de hacer en otros lenguajes. Supongamos que estuvi茅ramos haciendo una limpieza de datos y necesit谩ramos aplicar un mont贸n de transformaciones a la siguiente lista de cadenas:

Una forma de hacerlo es usar m茅todos de cadena integrados junto con el m贸dulo de biblioteca est谩ndar para expresiones regulares <code>re</code></p>




````python
import re
````

````python
def limpieza(palabras):
  resultado = []
  for i in palabras:
    i = i.strip()
    i = re.sub("[!#?]", "", i)
    i = i.title()
    resultado.append(i)
  return resultado
````

# **<font color="DeepPink">Generadores</font>**

<p align="justify">
Muchos objetos en Python admiten la iteraci贸n. Esto se logra mediante el protocolo iterador, una forma gen茅rica de hacer que los objetos sean iterables. <br><br>Por ejemplo, al iterar sobre un diccionario se obtienen las claves del diccionario:
</p>


````python
componentes
````

````python
for key in componentes:
  print(key)
````

<p align="justify">
 Un generador es una forma conveniente, similar a escribir una funci贸n normal, de construir un nuevo objeto iterable. <br><br>Mientras que las funciones normales ejecutan y devuelven un solo resultado a la vez, los generadores pueden devolver una secuencia de m煤ltiples valores pausando y reanudando la ejecuci贸n cada vez que se utiliza el generador. <br><br>Para crear un generador, utilice la palabra clave <code>yield</code> en lugar de <code>return</code>.
</p>


````python
def cuadrado(n=10):
  print(f"Generando cuadrados desde el valor 1 al valor {n ** 2}")
  for i in range(1, n + 1):
    yield i ** 2
````

<p align="justify">
 Cuando se llama al generador, no se ejecuta ning煤n c贸digo inmediatamente...
</p>


````python
generador = cuadrado()
generador
````

````python
type(generador)
````

<p align="justify">
 No es hasta que se solicita elementos del generador que comienza a ejecutarse su c贸digo.
</p>


````python
for i in generador:
  print(i, end=" ")
````

<p align="justify">
Dado que los generadores producen salida de un elemento a la vez en lugar de una lista completa a la vez, esto puede ayudar a usar menos memoria.</p>

Este es un generador an谩logo para enumerar, para diccionarios y tambien establecer comprensiones. <br><br>Para crear una expresi贸n generadora, se debe incluir una comprensi贸n de lista entre par茅ntesis en lugar de hacerla entre corchetes.<br><br>Por ejemplo:
</p>


````python
generador = (x ** 2 for x in range(10))
````

<p align="justify">
 Esto es equivalente al siguiente generador m谩s detallado:
</p>


````python
def generador():
  for x in range(10):
    yield x ** 2
generador = generador()
````

# **<font color="DeepPink">Manejo de errores y excepciones</font>**

<p align="justify">
Manejar los errores o excepciones de Python es una parte importante de la creaci贸n de scripts robustos. En las aplicaciones de an谩lisis de datos, muchas funciones solo funcionan en ciertos tipos de entrada. Ejemplo, la funci贸n <code>float()</code> que es capaz de convertir una cadena de caracteres a un n煤mero decimal, pero falla con entradas incorrectas.<br><br>Por ejemplo:

</p>


````python
float("45.90")
````

````python
float("numero")
````

<p align="justify">
Para evitar la falla por una entrada incorrecta,  generamos una funci贸n encerrando el bloque de la misma con <code>try</code> y <code>except</code>. Si la entrada de la funci贸n es correcta, entonces se ejecuta <code>try</code>, y si la entrada de la funci贸n no es la correcta, entonces se ejecuta <code>except</code>.<br><br>Por ejemplo:
</p>


````python
def a_decimal(x):
  try:
    return float(x)
  except:
    return x
````

````python
a_decimal("45.98")
````

````python
a_decimal("numero")
````

<p align="justify">
 Pero, se pueden generar excepciones de distintos  errores...
</p>


````python
float((298, 874))
````

````python
def a_decimal(x):
  try:
    return float(x)
  except (TypeError, ValueError):
    return x
````

````python
a_decimal((298, 874))
````

<p align="justify">
Entonces, las excepciones son errores detectados por Python durante la ejecuci贸n del programa. Cuando el int茅rprete se encuentra con una situaci贸n excepcional, como el intentar dividir un n煤mero entre 0 o el intentar acceder a un archivo que no existe, este genera o lanza una excepci贸n, informando al usuario de que existe alg煤n problema.

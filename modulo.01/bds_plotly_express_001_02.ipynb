 # **<font color="DeepPink">Usando plotly, gr치ficos lineales 游늴 </font>**

La biblioteca gr치fica <code>Plotly</code> crea gr치ficos interactivos con calidad de publicaci칩n. Ejemplos de c칩mo crear gr치ficos de l칤neas, diagramas de dispersi칩n, gr치ficos de 치rea, gr치ficos de barras, barras de error, diagramas de caja, histogramas, mapas de calor, subgr치ficos, ejes m칰ltiples, gr치ficos polares, gr치ficos de burbujas y todo tipos de gr치ficos, se desarrollar치n a continuaci칩n.

<p align="justify"> 游 Por convenci칩n, as칤 se importa <code>Plotly</code>:  </p>

```python
import plotly.express as px
```

<p align="justify">
Para poder graficar, vamos a utilizar un conjunto de datos correspondientes a un Fondo Com칰n de Inversi칩n</p>

```python
import pandas as pd
fci = pd.read_csv("https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/fci.csv")
fci.Fecha = pd.to_datetime(fci.Fecha,infer_datetime_format=True)
fci = fci.round(2)
fci.info()
```

 # **<font color="DeepPink">Seguimos con los gr치ficos lineales</font>**

<p align="justify">
游 Con <code>px.line()</code> cada punto de datos se representa como coordenadas (cuya ubicaci칩n proviene de la columna $x$ y la columna $y$) generando una l칤nea que une esas coordenadas, en un espacio bidimensional.<br><br> Antes de graficar, vemos los datos de un fondo com칰n de inversi칩n, filtrando por los rescates y las suscripciones:</p>

```python
fci.query("Concepto == 'Rescate'").head()
```

```python
fci.query("Concepto == 'Suscripci칩n'").head()
```

<p align="justify">
Entonces los gr치ficos l칤neales de <code>Plotly</code> se implementan como diagramas de dispersi칩n conectados, lo que significa que los puntos se trazan y se conectan con l칤neas en el orden en que se proporcionan, sin reordenaci칩n autom치tica.<br><br>Esto hace posible hacer gr치ficos como el siguiente (el que se encuentra ordenado por fechas), pero tambi칠n significa que puede ser necesario ordenar expl칤citamente los datos antes de graficarlos para evitar que las l칤neas se muevan "hacia atr치s" a trav칠s del gr치fico.
</p>



<p align="justify">
游 En el siguiente gr치fico, usamos algunos par치metros, tales como:</p>

- el dataframe.
- la serie asignada a $x$ (tambien puede ser una lista de series).
- la serie asignada a $y$.
- el t칤tulo del gr치fico.
- el template utilizado.
- los marcadores sobre la l칤nea.
- y los labels (para cambiar los nombres de las etiquetas).

```python
fig = px.line(fci.query("Concepto =='Rescate'"),
              x="Fecha",
              y="Valor_de_Cuota",
              title='Fondo Com칰n de Inversi칩n',
              template="gridon",
              markers=True,
              labels={"Valor_de_Cuota":"Valor de Cuota"})
fig.show()
```

<p align="justify">
游 En el siguiente gr치fico es lo mismo, pero se aplica otro filtro, para visualizar solo las suscripciones. Interesante destacar que se puede colocar dentro del parametro dataframe, un filtro del mismo.</p>


```python
fig = px.line(fci.query("Concepto =='Suscripci칩n'"),
              x="Fecha",
              y="Valor_de_Cuota",
              title='Fondo Com칰n de Inversi칩n',
              template="gridon",
              markers=True,
              labels={"Valor_de_Cuota":"Valor de Cuota"})
fig.show()
```

<p align="justify">
游 En el siguiente gr치fico excluye uno de los valores de la serie <code>Concepto</code>, y agrega un parametro denominado <code>color</code>, que incorpora una variable m치s a nuestro gr치fico.</p>


```python
px.line(fci.query("Concepto !='Saldo Anterior'"),
        x="Fecha",
        y="Valor_de_Cuota",
        color="Concepto",
        title='Fondo Com칰n de Inversi칩n',
        template="gridon",
        markers=True,
        labels={"Valor_de_Cuota":"Valor de Cuota"})
```

<p align="justify">
游 Ahora cambiamos el parametro <code>markers</code> por <code>symbol</code>, que agrega sobre la l칤nea, distintos s칤mbolos donde estaban los marcadores.</p>


```python
px.line(fci.query("Concepto !='Saldo Anterior'"),
        x="Fecha",
        y="Valor_de_Cuota",
        color="Concepto",
        title='Fondo Com칰n de Inversi칩n',
        template="gridon",
        symbol="Concepto",
        labels={"Valor_de_Cuota":"Valor de Cuota"})
```

<p align="justify">
游 Optimizando un poco mas el gr치fico, utilizamos <code>update_traces</code> para colocar sobre la l칤nea, el valor de cada punto. Hay varias posiciones del texto, en este caso, se establece que los valores se ubican por debajo de los puntos de la l칤nea, y a la derecha.</p>


```python
fig = px.line(fci.query("Concepto =='Suscripci칩n'"),
              x="Fecha",
              y="Valor_de_Cuota",
              title='Fondo Com칰n de Inversi칩n',
              template="gridon",
              markers=True,
              labels={"Valor_de_Cuota":"Valor de Cuota"},
              text="Valor_de_Cuota")

fig.update_traces(textposition="bottom right")

fig.show()
```

<p align="justify">
游 Ahora vamos adicionar botones para poder segmentar nuestro gr치fico, para ello hacemos un <code>update_xaxes()</code> con el par치metro <code>rangeselector</code> armando una lista para los botones, y cada boton es un diccionario.</p>


```python
fig = px.line(fci.query("Concepto =='Suscripci칩n'"),
              x="Fecha",
              y="Valor_de_Cuota",
              title='Fondo Com칰n de Inversi칩n',
              template="gridon",
              markers=True,
              labels={"Valor_de_Cuota":"Valor de Cuota"},
              text="Valor_de_Cuota")

fig.update_traces(textposition="bottom right")

fig.update_xaxes(rangeselector=dict(buttons=list([
    dict(count=7, label="7d", step="day", stepmode="backward"),
    dict(count=14, label="14d", step="day", stepmode="backward"),
    dict(count=1, label="1m", step="month", stepmode="backward"),
    dict(count=2, label="2m", step="month", stepmode="backward"),
    dict(label = "M치x.", step="all")])))

fig.show()
```

<p align="justify">
游 Ahora agregamos una l칤nea constante, con la media de los valores de suscripci칩n, para ello utilizamos <code>add_hline</code> que es una linea horizontal.</p>


```python
fci1 = fci.query("Concepto =='Suscripci칩n'")

fig = px.line(fci1,
              x="Fecha",
              y="Valor_de_Cuota",
              title='Fondo Com칰n de Inversi칩n',
              template="gridon",
              markers=True,
              labels={"Valor_de_Cuota":"Valor de Cuota"},
              text="Valor_de_Cuota")

fig.update_traces(textposition="bottom right")

fig.update_xaxes(rangeselector=dict(buttons=list([
    dict(count=7, label="7d", step="day", stepmode="backward"),
    dict(count=14, label="14d", step="day", stepmode="backward"),
    dict(count=1, label="1m", step="month", stepmode="backward"),
    dict(count=2, label="2m", step="month", stepmode="backward"),
    dict(label = "M치x.", step="all")])))

fig.add_hline(y=fci1.Valor_de_Cuota.mean(),
              line_color="green",
              line_dash="dot",
              name="Media")

fig.show()
```

<p align="justify">
游 Y verificamos el valor de la media, calculando con <code>mean()</code> para ver si la linea que adicionamos esta bien posicionada, y vemos que si.<br><br>El truco en este 칰ltimo grafico fue filtrar por fuera del gr치fico, asignar el resultante de ese filtrado a un objeto, y trabajar con ese objeto en el script del gr치fico</p>


```python
fci1.Valor_de_Cuota.mean().round(2)
```

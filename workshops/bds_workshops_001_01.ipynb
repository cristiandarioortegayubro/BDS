 ## **<font color="DeepPink">Grafico lineal</font>**

<p align="justify">  Queremos ver del <code>dataframe</code> las cotizaciones de cierre que estan en la columna <code>Close</code> </p>


```python
px.line(google, x="Date", y="Close")
```

```python
px.line(google, 
        x="Date", 
        y="Close",
        template="gridon",
        markers=True,
        title="Google 2022 - 2023")
```

```python
px.line(google.query("Date > '2023-01-01'"), 
        x="Date", 
        y="Close",
        template="gridon",
        markers=True,
        title="Google 2023")
```

<p align="justify">  Ahora supongamos que queremos ver del <code>dataframe</code> todas las cotizaciones estan en las columnas </p>


```python
google.columns[1:5]
```

```python
for i in google.columns[1:5]:
  fig = px.line(google.query("Date > '2023-01-01'"), 
        x="Date", 
        y=i,
        template="gridon",
        markers=True,
        title=f"Google 2023 - {i}")
  fig.show()
```

## **<font color="DeepPink">BoxPlot</font>**

```python
px.box(google.query("Date > '2023-01-01'"), 
       y="Close",
       template="gridon",
       points="all",
       title="Google 2023",
       hover_name="Date")
```

```python
px.box(google.query("Date > '2023-01-01'"), 
       y=google.columns[1:5],
       template="gridon",
       points="all",
       title="Google 2023",
       hover_name="Date")
```

 # **<font color="DeepPink">Limpieza y transformaci贸n de datos</font>**

<p align="justify">  Vamos a crear una columna en cada <code>dataframe</code> para asignar el nombre de cada acci贸n que se est谩 evaluando, con el objetivo de generar un conjunto de datos a lo largo, cuando se concatenen los tres <code>dataframe</code>:  </p>

```python
apple["Name"] = "AAPL"
google["Name"] = "GOOG"
microsoft["Name"] = "MSFT"
```

```python
apple.head()
```

```python
google.head()
```

```python
microsoft.head()
```

<p align="justify">  Ahora s铆 vamos a concatenar, para luego hacer todas las tareas que sean pertinente, una sola vez, es decir, en el <code>dataframe</code> concatenado:  </p>

```python
acciones = pd.concat([apple, google, microsoft])
```

```python
acciones.head()
```

<p align="justify">  Bueno, una vez concatenado, sacamos las columnas que no nos parecen pertinentes, como el cierre ajustado y el volumen, luego a ello, vamos a ordenar las columnas, dejando la columna <code>Name</code> en primer lugar...  </p>

<p align="justify">  Esta es una forma de hacerlo[texto del v铆nculo](https://):  </p>

```python
acciones.drop(columns=["Adj Close","Volume"])
```

<p align="justify">  Otra forma ser铆a, primero ordeno las columnas y luego filtro por columna con el m茅todo <code>iloc</code>:  </p>

```python
acciones.columns
```

```python
acciones = acciones[['Name','Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
```

```python
acciones.columns
```

```python
acciones.iloc[:,0:5]
```

```python
acciones = acciones.iloc[:,0:5]
```

```python
acciones.info()
```

 # **<font color="DeepPink">Graficando, otra vez, con Plotly</font>**

 ## **<font color="DeepPink">Grafico lineal</font>**

```python
acciones.columns
```

```python
for i in acciones.columns[1:5]:
  fig = px.line(acciones, 
        x=acciones.index, 
        y=i,
        color="Name",
        template="gridon",
        markers=True,
        title=f"Acciones de Apple, Google y Microsoft 2023 - {i}")
  fig.show()
```

## **<font color="DeepPink">Grafico de velas</font>**

```python
fig = go.Figure(data=[go.Candlestick(x=apple.index,
                open=apple['Open'],
                high=apple['High'],
                low=apple['Low'],
                close=apple['Close'])])

fig.update_layout(template="gridon",
                  title="Cotizaci贸n de Apple")


fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=google.index,
                open=google['Open'],
                high=google['High'],
                low=google['Low'],
                close=google['Close'])])

fig.update_layout(template="gridon",
                  title="Cotizaci贸n de Google")


fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=microsoft.index,
                open=microsoft['Open'],
                high=microsoft['High'],
                low=microsoft['Low'],
                close=microsoft['Close'])])

fig.update_layout(template="gridon",
                  title="Cotizaci贸n de Microsoft")


fig.show()
```

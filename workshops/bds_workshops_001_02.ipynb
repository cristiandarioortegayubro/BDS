 # **<font color="DeepPink">Limpieza y transformaci칩n de datos</font>**

<p align="justify"> 游 Vamos a crear una columna en cada <code>dataframe</code> para asignar el nombre de cada acci칩n que se est치 evaluando, con el objetivo de generar un conjunto de datos a lo largo, cuando se concatenen los tres <code>dataframe</code>:  </p>

```python
apple["Name"] = "Apple"
google["Name"] = "Google"
microsoft["Name"] = "Microsoft"
```

```python
apple.head()
```

```python
google.head()
```

```python
microsoft.head()
```

<p align="justify"> 游 Ahora s칤 vamos a concatenar, para luego hacer todas las tareas que sean pertinente, una sola vez, es decir, en el <code>dataframe</code> concatenado:  </p>

```python
acciones = pd.concat([apple, google, microsoft])
```

```python
acciones
```

<p align="justify"> 游 Bueno, una vez concatenado, sacamos las columnas que no nos parecen pertinentes, como el cierre ajustado y el volumen, luego a ello, vamos a ordenar las columnas, dejando la columna <code>Name</code> en primer lugar...  </p>

<p align="justify"> 游댄 Esta es una forma de hacerlo:  </p>

```python
acciones.drop(columns=["Adj Close","Volume"])
```

<p align="justify"> 游댄 Otra forma ser칤a, primero ordeno las columnas y luego dejo las columnas que quiero que queden en mi <code>dataframe</code>:  </p>

```python
acciones.columns
```

```python
acciones = acciones[['Name','Open', 'High', 'Low', 'Close']]
```

```python
acciones.columns
```

```python
acciones
```

```python
acciones.info()
```

```python
px.line(acciones, 
        x=acciones.index, 
        y="Close",
        template="gridon",
        color="Name",
        markers=True,
        title="Acciones de Apple, Google y Microsoft 2023")
```

## **<font color="DeepPink">Grafico de velas</font>**

```python
fig = go.Figure(data=[go.Candlestick(x=apple.index,
                open=apple['Open'],
                high=apple['High'],
                low=apple['Low'],
                close=apple['Close'])])

fig.update_layout(template="gridon",
                  title="Cotizaci칩n de Apple")


fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=google.index,
                open=google['Open'],
                high=google['High'],
                low=google['Low'],
                close=google['Close'])])

fig.update_layout(template="gridon",
                  title="Cotizaci칩n de Google")


fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=microsoft.index,
                open=microsoft['Open'],
                high=microsoft['High'],
                low=microsoft['Low'],
                close=microsoft['Close'])])

fig.update_layout(template="gridon",
                  title="Cotizaci칩n de Microsoft")


fig.show()
```

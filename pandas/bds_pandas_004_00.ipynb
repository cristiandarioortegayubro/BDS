 # **<font color="DeepPink">Manejo de datos faltantes</font>**

<p align="justify">
La falta de datos ocurre com煤nmente en muchas aplicaciones de an谩lisis de datos. Todas las estad铆sticas descriptivas de los objetos <code>Pandas</code> excluyen los datos faltantes de forma predeterminada. La forma en que se representan los datos faltantes en los objetos <code>Pandas</code> es algo imperfecta, pero suficiente para la mayor铆a de los casos del mundo real.</p>

<p align="justify">  Vamos a traer un conjunto de datos de la web.
</p>

```python
import lxml
import bs4
import html5lib
```

```python
URL = "https://atm.mendoza.gov.ar/portalatm/zoneBottom/datosInteres/recaudacion/recaudacion_impuesto_ingresos.jsp"
recaudacion = pd.read_html(URL)
df = recaudacion[0]
df
```

<p align="justify">  <code>Pandas</code> usa el valor de coma flotante <code>NaN</code>, que no es un n煤mero, para representar los datos que faltan. El m茅todo <code>isna()</code> nos da una serie booleana con <code>True</code> para los valores nulos.
</p>

```python
df.isna().head()
```

<p align="justify">  En <code>Pandas</code> se adopta por convenci贸n decirles a los datos faltantes, datos no disponibles, los <code>NA</code>. En  aplicaciones estad铆sticas, los datos no disponibles pueden ser datos que no existen o que existen pero no se observaron (debido a problemas con la recopilaci贸n de datos, por ejemplo un valor desconocido en una encuesta).<br><br> Al limpiar los datos para el an谩lisis, a menudo es importante estudiar los datos faltantes para identificar los problemas de recopilaci贸n de datos o posibles sesgos en los datos causados por los datos faltantes. <br><br>El valor integrado <code>None</code> de <code>Python</code> tambi茅n se trata como dato no disponible. <code>Pandas</code> ha intentado que el trabajo con datos no disponibles sea coherente en todos los tipos de datos.
</p>

<p align="justify">  Algunas funciones relacionadas con el manejo de datos no disponibles:
<ul>
<li><code>dropna()</code> filtra las etiquetas de los ejes en funci贸n de si faltan datos en los valores de cada etiqueta, con distintos umbrales para la cantidad de datos faltantes que se deben tolerar</li>
<li><code>fillna()</code> completa los datos faltantes con alg煤n valor predeterminado o utilizando alg煤n m茅todo</li>
<li><code>isna()</code> devuelve valores booleanos que indican qu茅 valores faltan</li>

<p align="justify"> En lugar de filtrar datos faltantes, y tal vez descartar otros datos junto con ese filtrado, es posible rellenarlos de varias maneras. El m茅todo <code>fillna()</code> es el m谩s utilizado en <code>Pandas</code>.  ## **<font color="DeepPink">Transformar datos usando una funci贸n o mapeo</font>**

<p align="justify"> Para muchos conjuntos de datos, es posible realizar alguna transformaci贸n basada en los valores de una matriz, serie o columna en un <code>DataFrame</code>. 

```python
df.dropna(axis=1, how="any", inplace=True)
df.head()
```

<p align="justify">  Supongamos que deseamos agregar una columna que indique con n煤meros el mes resultante de la columna Meses. Previo a ello, vamos a poner nombres de columnas y eliminar el indice cero.

```python
df.columns = df.iloc[1]
df.columns.name = None
df.drop([0,1], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

<p align="justify">  El m茅todo <code>map()</code> en una serie acepta una funci贸n o un objeto similar a un diccionario que contiene un mapeo para hacer la transformaci贸n de valores que estamos requiriendo.



```python
df.head()
```

```python
Mes = {"Enero":1, "Febrero":2, "Marzo":3, "Abril":4, "Mayo":5, "Junio":6,
       "Julio":7, "Agosto":8, "Septiembre":9, "Octubre":10, "Noviembre":11, "Diciembre":12}
```

```python
df.Meses = df.Meses.map(Mes)
df
```

<p align="justify">  Otra forma:




```python
df.head()
```

```python
Meses = {1:"Enero", 2:"Febrero", 3:"Marzo", 4:"Abril", 5:"Mayo", 6:"Junio",
         7:"Julio", 8:"Agosto", 9:"Septiembre", 10:"Octubre", 11:"Noviembre", 12:"Diciembre"}
```

```python
def meses(x):
  return Meses[x]
```

```python
df.Meses = df.Meses.map(meses)
df
```

<p align="justify">  Otra forma:


```python
df.head()
```

```python
Mes = {"Enero":1, "Febrero":2, "Marzo":3, "Abril":4, "Mayo":5, "Junio":6,
       "Julio":7, "Agosto":8, "Septiembre":9, "Octubre":10, "Noviembre":11, "Diciembre":12}
```

```python
df.Meses = df.Meses.map(lambda x: Mes[x])
df
```

 ## **<font color="DeepPink">Sustituci贸n de datos</font>**

<p align="justify"> Completar los datos que faltan con el m茅todo <code>fillna()</code> es un caso especial de reemplazo general. Como hemos visto, <code>map()</code> se puede usar para modificar un subconjunto de valores en un objeto, pero <code>replace()</code> proporciona una forma m谩s simple y flexible de hacerlo. 

```python
df.head()
```

```python
Meses = {1:"Enero", 2:"Febrero", 3:"Marzo", 4:"Abril", 5:"Mayo", 6:"Junio",
         7:"Julio", 8:"Agosto", 9:"Septiembre", 10:"Octubre", 11:"Noviembre", 12:"Diciembre"}
```

 ## **<font color="DeepPink">Discretizaci贸n</font>**

<p align="justify"> Los datos continuos a menudo se discretizan o se separan en "contenedores" para su an谩lisis. Supongamos que tenemos datos sobre un grupo de personas en un estudio comercial y deseamos agruparlos en grupos de edad discretos




```python
edad = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

<p align="justify">  Divid谩moslos en grupos de 18 a 25, de 26 a 35, de 36 a 60 y, finalmente, de 61 a帽os o m谩s. Para hacerlo, tenemos que usar <code>cut()</code>


```python
contenedores = [18, 25, 35, 60, 100]
```

```python
categorias = pd.cut(edad, contenedores)
```

```python
categorias
```

<p align="justify">  Cada contenedor se identifica mediante un tipo de valor de intervalo especial que contiene el l铆mite inferior y superior de cada contenedor


```python
pd.value_counts(categorias)
```

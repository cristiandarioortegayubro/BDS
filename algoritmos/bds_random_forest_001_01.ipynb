 # **<font color="DeepPink">Random Forest, Clasificaci贸n</font>**

<p align="justify">
 Recordando, un bosque aleatorio es un algoritmo de aprendizaje autom谩tico que se utiliza para la clasificaci贸n y la regresi贸n. Es un conjunto de 谩rboles de decisi贸n que se entrenan de forma independiente y luego se combinan para hacer una predicci贸n.


 # **<font color="DeepPink">Datos</font>**

```python
import pandas as pd
import numpy as np
```

```python
datos = "https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/income.csv"
```

```python
income = pd.read_csv(datos)
```

```python
income
```

```python
income.info()
```

```python
target_name = "income >50K"
target = income[target_name]
data = income.drop(columns=[target_name])
```

 # **<font color="DeepPink">Ajuste del modelo</font>**

 # **<font color="DeepPink">Preprocesamiento de los datos</font>**


```python
from sklearn.compose import make_column_selector as selector
```

```python
numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)
```

```python
numerical_columns = numerical_columns_selector(data)
categorical_columns = categorical_columns_selector(data)
```

```python
numerical_columns
```

```python
categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
categorical_preprocessor = OneHotEncoder(handle_unknown="ignore")
numerical_preprocessor = StandardScaler()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer([
    ('one-hot-encoder', categorical_preprocessor, categorical_columns),
    ('standard_scaler', numerical_preprocessor, numerical_columns)])
```

```python
preprocessor
```

 # **<font color="DeepPink">Creaci贸n del modelo</font>**


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(preprocessor, RandomForestClassifier(random_state = 123))
model
```

 # **<font color="DeepPink">Train-test, divisi贸n del conjunto de datos</font>**

```python
from sklearn.model_selection import train_test_split
```

```python
data_train, data_test, target_train, target_test = train_test_split(data,
                                                                    target,
                                                                    random_state=123)
```

 # **<font color="DeepPink">Ajuste y prediccion</font>**

```python
model.fit(data_train, target_train)
```

```python
model.score(data_test, target_test)
```

 # **<font color="DeepPink">Matriz de confusi贸n</font>**

<p align="justify">
La matriz de confusi贸n es una herramienta com煤nmente utilizada para evaluar el rendimiento de un modelo de clasificaci贸n, especialmente cuando se trata de problemas de dos clases, como la clasificaci贸n binaria, donde solo hay dos clases posibles: la clase 0 y la clase 1.
<br>
<br>
En un modelo de clasificaci贸n binaria, los resultados de las predicciones se pueden resumir en cuatro t茅rminos:

<ol align="justify">
<li><b>Verdaderos positivos (TP)</b>: Representa el n煤mero de muestras que pertenecen a la clase 1 (positiva) y fueron clasificadas correctamente como clase 1.</li><br>
<li><b>Verdaderos negativos (TN)</b>: Representa el n煤mero de muestras que pertenecen a la clase 0 (negativa) y fueron clasificadas correctamente como clase 0.</li><br>
<li><b>Falsos positivos (FP)</b>: Representa el n煤mero de muestras que pertenecen a la clase 0 (negativa) pero fueron incorrectamente clasificadas como clase 1.</li><br>
<li><b>Falsos negativos (FN)</b>: Representa el n煤mero de muestras que pertenecen a la clase 1 (positiva) pero fueron incorrectamente clasificadas como clase 0.</li>

```python
from sklearn.metrics import confusion_matrix
```

```python
mat_confusion = confusion_matrix(y_true = target_test,
                                 y_pred = model.predict(data_test))
```

```python
mat_confusion
```

```python
import plotly.express as px
```

```python
fig = px.imshow(mat_confusion,
                text_auto=True,
                labels=dict(x="Predict", y="True", color="Muestras"),
                x=['0', '1'],
                y=['0 ', '1 '])

fig.update_coloraxes(showscale=False)

fig.show()
```

 # **<font color="DeepPink">Reporte</font>**

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_true    = target_test,
                               y_pred    = model.predict(data_test)
                               )
```

```python
print(report)
```

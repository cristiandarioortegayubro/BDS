 # **<font color="DeepPink">Manejo de fechas con Python y Pandas</font>**

<p align="justify">
Las series temporales son puntos de datos recopilados a lo largo del tiempo. Por lo general, el tiempo est치 espaciado de manera uniforme entre cada punto de datos. Sin embargo, puede haber brechas en las observaciones. Pandas incluye funcionalidades para manipular fechas, realizar agregaciones en diferentes per칤odos de tiempo, muestrear diferentes intervalos de tiempo y m치s.
<br><br>
Antes de meternos con Pandas y como trabajar las series veremos el objeto principal que debemos aprender a manipular que es el datetime. En realidad se pueden tener tres tipos de  objetos que son <code>date</code>, <code>time</code> y <code>datetime</code>.
<br><br>
Por otro lado, Pandas tiene encapsulado los <code>date</code> y <code>time</code> en un tipo de objeto denominado <code>Timestamp</code>. Y con Numpy podemos ver los objetos datetime64 como tipo de datos, sin embargo todos tienen funciones similares.



```python
import pandas as pd
import numpy as np
import datetime
```

```python
date = datetime.date(year = 2023, month = 8, day = 20)
time = datetime.time(hour = 12, minute = 30, second = 19, microsecond = 463198)
dt = datetime.datetime(year = 2023, month = 8, day = 20, hour = 12, minute = 30, second = 18)

print(f"La fecha es {date}")
```

```python
print(f"La hora es {time}")
```

```python
print(f"la fecha y hora es {dt}")
```

游 Tambi칠n podemos construir lapsos de tiempo con el objeto <code>timedelta</code>

```python
td = datetime.timedelta(weeks = 2, days = 5, hours = 10 , minutes = 20)
td
```

游 Este objeto ahora se lo podemos sumar a cualquier fecha que ya tengamos.

```python
print(f"La nueva fecha es {dt+td}")
```

<p align="justify">
Vamos a dirigirnos a pandas y su objeto <code>Timestamp</code>, que es un momento en el tiempo con una precisi칩n de nanosegundos. El constructor de Timestamp es muy flexible y maneja una amplia variedad de entradas.

```python
pd.Timestamp(year = 2023, month = 2, day = 3, hour = 5, minute = 10)
```

<p align="justify">
Pandas tambien tiene la funci칩n <code>to_Timestamp</code> que tambi칠n tiene alta flexibilidad con algunas funciones m치s. Y esta funci칩n puede convertir series enteras a datos de tipo datetime

```python
s = pd.Series(['12-5-2015', '14-1-2013','20/12/2017', '40/23/2017'])
pd.to_datetime(s, dayfirst = True, errors = 'coerce')
```

<p align="justify">
Al igual que el constructor de <code>Timestamp</code> y la funci칩n <code>to_datetime</code>, pandas cuenta con <code>Timedelta</code> y <code>to_timedelta</code> para representar una cantidad de tiempo. Tanto el constructor <code>Timedelta</code> como la funci칩n <code>to_timedelta</code> pueden crear un objeto <code>Timedelta</code> 칰nico.
<br><br>
Similar a <code>to_datetime</code>, <code>to_timedelta</code> tiene un poco m치s de funcionalidad y puede convertir listas completas o Series en objetos Timedelta:

```python
 pd.Timedelta('12 days 5 hours 3 minutes 123456789 nanoseconds')
```

```python
 pd.Timedelta(days=5, minutes=7.34)
```

```python
s = pd.Series([10, 100])

pd.to_timedelta(s, unit='s')
```

```python
pd.to_timedelta(s, unit = 'd')
```

 ## **<font color="DeepPink">Seleccionando periodos</font>**


<p align="justify">
De la manera que venimos viendo podemos seleccionar periodos y agrupar pero cuando el 칤ndice de nuestra serie pasa a ser un <code>DatetimeIndex </code> hay mejores oportunidades para seleccionar y se realiza de manera m치s eficiente.
<br><br>
Vamos a leer un dataset que corresponde a pagos electr칩nicos capturados en los 칰ltimso meses.


```python
url = "https://raw.githubusercontent.com/pabanib/dataframes/master/pagos.csv"
pagos = pd.read_csv(url, sep = ';')
pagos.dtypes
```

```python
pagos.head()
```

<p align="justify">
Como vemos, nuestros datos no han interpretado las fechas y horas como si fuera un <code>datetime</code> por lo tanto vamos a tener que transformarlas

```python
pagos.ImpoPaon = pagos.ImpoPaon.apply(lambda x: float(x.replace(',','.')))
```

```python

for k, t in pagos.dtypes.items():
  if t == 'object':
    try:
      s = pd.to_datetime(pagos[k], dayfirst = True)
      pagos[k] = s
    except:
      pass
pagos.dtypes
```

```python
pagos.head()
```

<p align="justify">
Podemos cambiar el 칤ndice del df para llegar a acceder a mejores formas de trabajo.

```python
pagos = pagos.set_index('FechPaon')
pagos.head()
```

<p align="justify">
Podemos ubicar una fecha precisa psando la fehca y hora correcta

```python
pagos.loc['2023-07-02 00:00:00']
```

<p align="justify">
O podemos buscar todos lo movimientos en una fecha, mes, a침o, etc.

```python
pagos.loc['2023-07-02']
```

```python
pagos.loc['2023-07']
```

<p align="justify">
Adicionalmente se puede seleccionar un rango de fechas utilizando los <code>:</code> entre las fechas que deseamos.

```python
pagos.loc['2023-07-06':'2023-07-10'].sort_index()
```

<p align="justify">
Si queremos trabajar con la serie <code>datetime</code> sin que sea un 칤ndice tiene otras complicaci칩nes y no es tan accesible como un 칤ndice que sea temporal.
<br><br>
Si ponemos la fecha y hora compleata lo filtra.

```python
pagos[pagos.AltaPago == '2023-07-10 22:54:00']
```

<p align="justify">
Pero no sucede en caso de que yo quiera buscar los registros espec칤ficos para ese d칤a.

```python
pagos[pagos.AltaPago == '2023-07-10']
```

<p align="justify">
Y tampoco filtra nada si queremos usar el atributo <code>.dt</code>.

```python
pagos[pagos.AltaPago.dt.date == '2023-07-10']
```

<p align="justify">
Solo funcionar칤a si tratamos de comparar con  un objeto <code>datetime.date</code>

```python
pagos[pagos.AltaPago.dt.date == datetime.date(year = 2023, month = 7, day = 10)]
```

<p align="justify">
Pero si podemos trabajar con el m칠todo <code>between</code> que nos permite elegir entre dos fechas.

```python
pagos[pagos.AltaPago.between('2023-07-10','2023-07-10 01:00')]
```

 ## **<font color="DeepPink">Agrupando por per칤odos de tiempos equidistantes</font>**


<p align="justify">
Con series temporales se puede usar el m칠todo <code>.groupby</code> incluido en cualquier DataFrame de pandas. Pero existe otro m칠todo adem치s que funciona para agrupar cuando tenemos fechas que se denomina <code>.resample</code>.
<br><br>
Si quisieramos agrupar por d칤a el dataframe, con groupby deber칤amos pasarles los d칤as, que pasar칤an a ser los nuevos 칤ndices. Con resample simplemente le pasamos el argumento 'd'.

```python
pagos.groupby(pagos.index.date).count()
```

```python
pagos.resample('d').count()
```

```python
pagos.resample('12h').sum()['ImpoPaon'].plot()
```

```python
pagos.resample('W').count()[['ImpoPaon']].T.plot(kind = 'bar', figsize = (8,6))
```

Tambi칠n podemos pasar diferentes funciones a las diferentes columnas con el m칠todo <code>.agg</code> como con los <code>groupby</code>.

```python
pagos.resample('W').agg({'CodiPaon':'count', 'ImpoPaon':'sum'})
```

 ## **<font color="DeepPink">Otras formas de agrupar los datos</font>**


<p align="justify">
Tal vez estemos interesados en saber en que hora del d칤a es cuando se producen m치s cantidad de pagos. Bueno para realizarlo podemos utilizar <code>groupby</code> y luego le pasamos las horas de nuestro 칤ndice con el atributo <code>.hours</code>

```python
pagos.groupby(pagos.index.hour).size().plot(kind = 'bar')
```

<p align="justify">
Al parecer tenemos un error con la hora 0, probablemente son valores que vinieron sin hora espec칤fica. Podemos tambi칠n agrupar con alguna serie del dataframe que sea temporal, pero cuando no es 칤ndice necesitamos utilizar primero el atributo <code>.dt</code>.

```python
pagos.groupby(pagos.AltaPago.dt.hour).size().plot(kind = 'bar')
```

<p align="justify">
Tambi칠n podemos saber en que d칤a de la semana se produce el evento. Para esto podemos usar el m칠todo <code>.day_name</code>.

```python
pagos.groupby(pagos.index.day_name()).size().plot(kind = 'bar')
```

Por supuesto tambi칠n lo podemos hacer con una serie

```python
pagos['day'] = pagos.AltaPago.dt.day_of_week
pagos.groupby(pagos.AltaPago.dt.day_name()).agg({'day':'first', 'ImpoPaon':'count'}).sort_values('day')['ImpoPaon'].plot(kind = 'bar')
```

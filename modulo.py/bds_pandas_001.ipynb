Lo que vamos a hacer ahora, es cambiar ese formato numérico científico..., para ello utilizamos de pandas, opciones y configuraciones...

## Opciones y configuraciones

https://pandas.pydata.org/docs/user_guide/options.html

Pandas puede configurar el comportamiento global de los dataframe, y su visualización.

con el método ```describe_option()``` se puede ver la configuración actual... Si dentro de los paréntesis no se establece un parámetro, el método entonces muestra todas las opciones y configuraciones disponibles... 

Por el contrario, si como parámetro se establece una opción, entonces se podrá acceder a todas las configuraciones posibles de esa opción, como lo sería en el caso del parámetro ```display``` 

````python
pd.describe_option('display')
````

### Formato cientifico y cantidad de decimales

```display.precision``` se utiliza para configurar la cantidad de decimales...

````python
pd.describe_option('display.precision')
````

```float_format``` establece el comportamiento de los decimales, y cambia el aspecto científico de los números...

````python
pd.describe_option('float_format')
````

````python
pd.options.display.precision = 2
````

````python
df1.describe()
````

````python
pd.options.display.float_format = "{:.2f}".format
````

````python
df1.describe()
````

### Cantidad de filas del dataframe

````python
pd.describe_option('display.max_rows')
````

````python
pd.describe_option('display.min_rows')
````

````python
df1.shape
````

````python
pd.options.display.max_rows = 62
````

````python
df1
````

...y para volver a los parámetros iniciales... se utiliza el método ```reset_option()``` y entre paréntesis se coloca la opción que se quiere regresar a los valores por defecto...

````python
pd.reset_option("display.max_rows")
````

````python
df1
````

````python
pd.options.display.max_rows = None
````

````python
df1
````

````python
pd.reset_option("display.max_rows")
````

...si lo que se quiere es ver mas de cinco filas iniciales y finales cuando se visualiza el dataframe, entonces se utiliza ```display.min_rows```...

Ejemplo, tres filas iniciales y tres finales...

````python
pd.options.display.min_rows = 6
````

````python
df1
````

````python
pd.reset_option("display.min_rows")
````

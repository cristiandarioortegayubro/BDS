 # **<font color="DeepPink">PyCaret - Clustering</font>**

<p align="justify">
En <code>PyCaret</code>, el m√≥dulo de Clustering (Clustering Module) es una parte esencial de la biblioteca dise√±ada para realizar tareas de agrupamiento automatizado. El agrupamiento, tambi√©n conocido como clustering en ingl√©s, es una t√©cnica de aprendizaje autom√°tico no supervisado que tiene como objetivo dividir un conjunto de datos en grupos o cl√∫steres, donde los elementos dentro de cada cl√∫ster son similares entre s√≠ y diferentes de los elementos en otros cl√∫steres.
<br><br>
El m√≥dulo de Clustering de <code>PyCaret</code> ofrece una serie de funciones y herramientas que facilitan el proceso de agrupamiento de datos de manera automatizada y eficiente.
<br><br>
Algunas de las caracter√≠sticas clave del m√≥dulo de Clustering de <code>PyCaret</code> son:
<br>
<br>
<ol align="justify">
<li><b>Selecci√≥n de algoritmo</b>: <code>PyCaret</code> admite una variedad de algoritmos de agrupamiento populares, como K-Means, DBSCAN, Mean Shift, y m√°s. Los usuarios pueden elegir entre diferentes algoritmos para encontrar el que mejor se ajuste a sus datos y objetivos.</li>
<br>
<li><b>Optimizaci√≥n de hiperpar√°metros</b>: Al igual que en los m√≥dulos de Clasificaci√≥n y Regresi√≥n, <code>PyCaret</code> realiza una b√∫squeda inteligente en el espacio de hiperpar√°metros para encontrar la configuraci√≥n √≥ptima para clustering.</li>
<br>
<li><b>Visualizaci√≥n de cl√∫steres</b>: <code>PyCaret</code> proporciona herramientas para visualizar los resultados del agrupamiento, lo que ayuda a comprender c√≥mo se han formado los cl√∫steres y c√≥mo se distribuyen los datos dentro de ellos.</li>
<br>
<li><b>An√°lisis y selecci√≥n de caracter√≠sticas</b>: <code>PyCaret</code> permite a los usuarios realizar an√°lisis y selecci√≥n de caracter√≠sticas antes de aplicar el algoritmo de agrupamiento, lo que ayuda a mejorar la calidad y la interpretaci√≥n de los resultados.</li>
<br>
<li><b>Evaluaci√≥n del agrupamiento</b>: <code>PyCaret</code> proporciona m√©tricas de evaluaci√≥n del agrupamiento, como  Silueta (Silhouette) y la Inercia (Inertia), que ayudan a evaluar la calidad de los cl√∫steres generados.</li>
<br>
<li><b>Predicci√≥n en nuevos datos</b>: Una vez que se ha realizado el agrupamiento en el conjunto de entrenamiento, PyCaret permite aplicar el modelo de agrupamiento a nuevos datos para asignarlos a los cl√∫steres existentes.</li>
</ol>
<br>
<p align="justify">
El m√≥dulo de Clustering de <code>PyCaret</code> es una herramienta valiosa para estudiantes y profesionales de Ciencia de Datos que deseen explorar y comprender patrones y estructuras ocultas en sus datos sin la necesidad de supervisi√≥n. Al automatizar gran parte del proceso de agrupamiento, <code>PyCaret</code> permite a los usuarios centrarse en el an√°lisis e interpretaci√≥n de los resultados, lo que facilita la toma de decisiones informadas y la obtenci√≥n de informaci√≥n valiosa de los datos no etiquetados.

 ### **<font color="DeepPink">Silhouette</font>**

<p align="justify">
El coeficiente de Silhouette, com√∫nmente conocido como "Silhouette", es una m√©trica utilizada para evaluar la calidad de un  cluster de datos. Se utiliza en t√©cnicas de aprendizaje no supervisado, como el clustering, para medir qu√© tan bien los datos est√°n agrupados y separados unos de otros.
<br><br>
El coeficiente de Silhouette se calcula para cada punto de datos y proporciona una puntuaci√≥n que va desde $-1$ hasta $1$. Los valores m√°s altos de Silhouette indican que un punto de datos est√° bien agrupado, mientras que valores cercanos a $-1$ indican que el punto de datos podr√≠a haber sido asignado a un grupo incorrecto.
<br><br>
El c√°lculo del coeficiente de Silhouette implica tres componentes principales:
<br><br>
<ol align="justify">
<li><b>Distancia intra-cluster (a)</b>: Es la media de la distancia entre un punto y todos los dem√°s puntos en el mismo cluster al que pertenece. Representa qu√© tan compacto est√° el punto dentro de su cluster.</li>
<br>
<li><b>Distancia media al cluster m√°s cercano (b)</b>: Es la media de la distancia entre un punto y todos los puntos en el cluster m√°s cercano diferente al que pertenece. Representa qu√© tan separado est√° el punto de otros clusters.</li>
<br>
<li><b>Coeficiente de Silhouette (s)</b>: Se calcula para cada punto usando la siguiente f√≥rmula:</li>
<br><br>
 $$s = \frac{b - a}{\max(a, b)}$$
<br>
Una vez calculados los coeficientes de Silhouette para todos los puntos, se puede tomar el promedio de todos los coeficientes para obtener la puntuaci√≥n general del agrupamiento.</ol>
<br>
<p align="justify">
Interpretaci√≥n del coeficiente de Silhouette:
<br>
<ul align="justify">
<li> Si el coeficiente de Silhouette es cercano a $1$, significa que el punto est√° bien agrupado y separado de otros clusters, lo que indica un buen agrupamiento.</li>
<br>
<li> Si el coeficiente de Silhouette es cercano a $0$, indica que el punto est√° cerca del l√≠mite entre dos clusters y podr√≠a haber sido asignado a diferentes clusters, lo que sugiere una ambig√ºedad en la agrupaci√≥n.</li>
<br>
<li> Si el coeficiente de Silhouette es cercano a $-1$, indica que el punto est√° m√°s cerca de un cluster diferente al que pertenece, lo que sugiere un mal agrupamiento.</li></ul><br>
<p align="justify">
En resumen, el coeficiente de Silhouette es una medida √∫til para evaluar la calidad del agrupamiento de datos y seleccionar el n√∫mero √≥ptimo de clusters en t√©cnicas de clustering.
<br><br>
Un valor alto de Silhouette indica que el agrupamiento es m√°s s√≥lido y confiable, mientras que un valor bajo sugiere que puede haber problemas con la separaci√≥n de los datos en grupos coherentes.

 ### **<font color="DeepPink">Calinski-Harabasz</font>**

<p align="justify">
El √≠ndice de Calinski-Harabasz, tambi√©n conocido como √≠ndice de varianza entre grupos (Variance Ratio Criterion), es una m√©trica utilizada para evaluar la calidad de un agrupamiento (cluster) de datos en t√©cnicas de aprendizaje no supervisado, como el clustering.
<br><br>
El √≠ndice de Calinski-Harabasz se basa en la idea de que un buen agrupamiento deber√≠a tener una alta dispersi√≥n entre los grupos (clusters) y una baja dispersi√≥n dentro de los grupos. Es decir, los puntos dentro de cada grupo deber√≠an estar bien agrupados y separados claramente de los puntos en otros grupos.
<br><br>
El c√°lculo del √≠ndice de Calinski-Harabasz implica tres componentes principales:
<br><br>
<ol align="justify">
<li><b>Dispersiones entre grupos (Between-Cluster Dispersion, B)</b>: Es una medida de qu√© tan separados est√°n los grupos entre s√≠. Se calcula sumando la dispersi√≥n entre el centroide global de los datos y los centroides de cada grupo, y luego multiplic√°ndolo por el n√∫mero de puntos en el conjunto de datos.</li>
<br>
<li><b>Dispersiones dentro de grupos (Within-Cluster Dispersion, W)</b>: Es una medida de qu√© tan compactos est√°n los puntos dentro de cada grupo. Se calcula sumando la dispersi√≥n dentro de cada grupo, que es la suma de las distancias entre cada punto y el centroide del grupo al que pertenece.</li>
<br>
<li><b>√çndice de Calinski-Harabasz (CH)</b>: Se calcula usando la siguiente f√≥rmula:</li>
<br>
 $$CH = \frac{B}{W} \times \frac{N - K}{K - 1}$$
<br>
Donde:
<br><ul align="justify">
<li> \( N \) es el n√∫mero total de puntos en el conjunto de datos.</li>
<li> \( K \) es el n√∫mero de clusters.</li>
</ul></ol>
<p align="justify">
El √≠ndice de Calinski-Harabasz busca maximizar la relaci√≥n entre la dispersi√≥n entre grupos y la dispersi√≥n dentro de los grupos. Cuanto mayor sea el valor de CH, mejor ser√° la calidad del agrupamiento. Un valor alto de CH indica que los grupos est√°n bien separados y compactos, lo que sugiere que el agrupamiento es s√≥lido y coherente.
<br>
Es importante tener en cuenta que el √≠ndice de Calinski-Harabasz puede ser sensible al tama√±o y forma de los clusters, por lo que debe utilizarse en conjunto con otras m√©tricas y t√©cnicas para evaluar y validar el agrupamiento en un problema espec√≠fico.
<br><br>
En general, el √≠ndice de Calinski-Harabasz es una herramienta √∫til para seleccionar el n√∫mero √≥ptimo de clusters y evaluar la calidad del agrupamiento en t√©cnicas de clustering.

 ### **<font color="DeepPink">Davies-Bouldin</font>**

<p align="justify">
El √≠ndice Davies-Bouldin es otra m√©trica utilizada para evaluar la calidad de un  cluster de datos en t√©cnicas de aprendizaje no supervisado, como el clustering. Al igual que el √≠ndice de Calinski-Harabasz, el √≠ndice Davies-Bouldin busca medir qu√© tan bien est√°n agrupados los datos y cu√°n separados est√°n los grupos entre s√≠.
<br><br>
El c√°lculo del √≠ndice Davies-Bouldin se basa en la idea de que un buen agrupamiento deber√≠a tener grupos bien definidos y claramente separados entre s√≠.
<br><br>
Para calcular el √≠ndice, se realiza lo siguiente para cada par de clusters $i$ y $j$:
<br><br>
<ol align="justify">
<li>Se calcula la dispersi√≥n dentro del cluster \( i \) (denotada como \( d_i \)) como la distancia promedio entre cada punto en el cluster \( i \) y el centroide del mismo.</li>
<br>
<li>Se calcula la dispersi√≥n dentro del cluster \( j \) (denotada como \( d_j \)) de manera similar.</li>
<br>
<li>Se calcula la distancia entre los centroides de los clusters \( i \) y \( j \) (denotada como \( d_{ij} \)).</li>
<br>
<li>Se calcula el √≠ndice Davies-Bouldin para el par de clusters \( i \) y \( j \) como:
<br><br>
\[ R_{ij} = \frac{d_i + d_j}{d_{ij}} \]
<br></li>
<li>Se repite este proceso para todas las combinaciones de clusters y se obtiene el valor promedio de \( R_{ij} \) para todos los pares de clusters. El valor promedio se toma como el √≠ndice Davies-Bouldin para el agrupamiento completo.</li></ol>
<br><p align="justify">
Un √≠ndice Davies-Bouldin m√°s bajo indica un mejor agrupamiento, ya que significa que los grupos est√°n m√°s claramente separados y m√°s compactos dentro de s√≠ mismos. Por lo tanto, al comparar diferentes agrupamientos, aquellos que tengan un √≠ndice Davies-Bouldin m√°s bajo se consideran m√°s deseables.
<br><br>
Al igual que con otras m√©tricas de evaluaci√≥n de clustering, el √≠ndice Davies-Bouldin debe utilizarse junto con otras m√©tricas y t√©cnicas para obtener una comprensi√≥n completa de la calidad del agrupamiento y seleccionar el n√∫mero √≥ptimo de clusters en funci√≥n de la estructura de los datos.

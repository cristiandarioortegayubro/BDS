 # **<font color="DeepPink">PyCaret - Regresin - Optimizacin - Ensemble</font>**

<p align="justify">
En <code>PyCaret</code>, el mdulo de Regresin (Regression Module) es una parte importante de la biblioteca diseada para abordar tareas de regresin automatizada. La regresin es un tipo de problema de aprendizaje automtico supervisado en el que el objetivo es predecir un valor numrico continuo, en lugar de una clase o categora como en la clasificacin.
<br><br>
El mdulo de Regresin de <code>PyCaret</code> ofrece una serie de funciones y herramientas que simplifican el proceso de construccin, entrenamiento y evaluacin de modelos de regresin de manera automatizada y eficiente.
<br><br>
Algunas de las caractersticas clave del mdulo de Regresin de <code>PyCaret</code> son:
<br>
<br>
<ol align="justify">
<li><b>Preprocesamiento de datos</b>: <code>PyCaret</code> automatiza el preprocesamiento de datos para tareas de regresin, lo que incluye el manejo de valores faltantes, codificacin de variables categricas, normalizacin de datos, y particin de datos en conjuntos de entrenamiento y prueba. Esto permite a los usuarios concentrarse ms en el desarrollo del modelo y menos en las tareas de limpieza y transformacin de datos.</li>
<br>
<li><b>Comparacin de modelos</b>: <code>PyCaret</code> permite a los usuarios comparar fcilmente varios modelos de regresin populares. La biblioteca entrena automticamente mltiples modelos en el conjunto de datos y proporciona una tabla de resultados que muestra las mtricas de rendimiento de cada modelo, lo que facilita la identificacin del modelo ms prometedor para un problema en particular.</li>
<br>
<li><b>Optimizacin de hiperparmetros</b>: Los hiperparmetros son configuraciones que afectan el rendimiento y comportamiento de un modelo de aprendizaje automtico. <code>PyCaret</code> realiza una bsqueda inteligente en el espacio de hiperparmetros para encontrar la combinacin ptima que maximice el rendimiento del modelo de regresin.</li>
<br>
<li><b>Interpretacin de modelos</b>: <code>PyCaret</code> ofrece herramientas para interpretar y visualizar los modelos de regresin, lo que ayuda a comprender cmo se ajustan los datos y cmo se relacionan las caractersticas con las predicciones.</li>
<br>
<li><b>Anlisis de residuos</b>: Los residuos son las diferencias entre las predicciones del modelo y los valores reales. <code>PyCaret</code> proporciona herramientas para analizar y visualizar los residuos, lo que ayuda a evaluar la calidad del modelo.</li>

<li><b>Prediccin en nuevos datos</b>: Una vez que se ha construido y entrenado un modelo de regresin, <code>PyCaret</code> permite utilizarlo para hacer predicciones en nuevos datos sin la necesidad de escribir mucho cdigo adicional.</li>

 ## **<font color="DeepPink">Metricas</font>**

 ### **<font color="DeepPink">MAE</font>**

<p align = "justify">
<b>MAE</b> (Error Absoluto Medio, por sus siglas en ingls) es una mtrica utilizada para evaluar el rendimiento de un modelo de regresin. Representa el promedio de las diferencias absolutas entre las predicciones del modelo y los valores reales de los datos de prueba.
<br><br>
Para calcular MAE, primero se deben hacer predicciones con el modelo sobre un conjunto de datos de prueba y luego se resta el valor real de cada dato de prueba a su correspondiente prediccin para obtener la diferencia absoluta. Despus, se toma el promedio de todas estas diferencias absolutas para obtener MAE.
<br><br>
La frmula para calcular MAE es la siguiente:
<br><br>
MAE $= \frac{1}{n} \sum_{i=1}^{n} |y_{i} - \hat{y_{i}}|$
<br><br>
Donde:
<br><br>
<ul align = "justify">
<li> \( n \) es el nmero de muestras en el conjunto de datos de prueba.</li><br>
<li> \( y_{i} \) es el valor real del dato \( i \) en el conjunto de datos de prueba.</li><br>
<li> \( \hat{y_{i}} \) es la prediccin del modelo para el dato \( i \) en el conjunto de datos de prueba.</li><br>
<li> \( |x| \) representa el valor absoluto de \( x \).</li><br>
</ul>
<p align="justify">
<b>MAE</b> es una medida de la magnitud promedio de los errores de prediccin del modelo, independientemente de la direccin de los errores (es decir, no toma en cuenta si el modelo subestima o sobreestima los valores).
<br><br>
Un MAE ms bajo indica un mejor rendimiento del modelo, ya que significa que las predicciones estn ms cercanas a los valores reales en promedio.

 ### **<font color="DeepPink">RMSE</font>**

<p align="justify">
<b>RMSE</b> (Root Mean Squared Error, por sus siglas en ingls) es otra mtrica comnmente utilizada para evaluar el rendimiento de un modelo de regresin. Al igual que el MAE, el RMSE mide la precisin del modelo al comparar las predicciones con los valores reales de los datos de prueba.
<br><br>
Sin embargo, a diferencia del MAE, el RMSE penaliza de manera ms significativa los errores grandes.
<br><br>
El RMSE se calcula de la siguiente manera:
<br><br>
RMSE = $\sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_{i} - \hat{y_{i}})^{2}}$
<br><br>
Donde:
<br><br>
<ul align="justify">
<li> \( n \) es el nmero de muestras en el conjunto de datos de prueba.</li><br>
<li> \( y_{i} \) es el valor real del dato \( i \) en el conjunto de datos de prueba.</li><br>
<li> \( \hat{y_{i}} \) es la prediccin del modelo para el dato \( i \) en el conjunto de datos de prueba.</li><br>
</ul>
<p align="justify">
RMSE toma la raz cuadrada del promedio de los errores al cuadrado. Al elevar los errores al cuadrado, los errores ms grandes se vuelven ms prominentes y afectan ms la mtrica, lo que penaliza fuertemente las predicciones que estn lejos de los valores reales.
<br><br>
Una vez ms, al igual que con el MAE, un RMSE ms bajo indica un mejor rendimiento del modelo, ya que significa que las predicciones estn ms cercanas a los valores reales en promedio.
<br><br>
El RMSE es especialmente til cuando se quiere tener en cuenta la magnitud de los errores y se quiere penalizar ms fuertemente las predicciones inexactas.

 ### **<font color="DeepPink">$R^2$</font>**

<p align="justify">
El coeficiente de determinacin, comnmente conocido como <b>$R^2$</b> (pronunciado "R cuadrado"), es una mtrica utilizada para evaluar la calidad de ajuste de un modelo de regresin.
<br><br>
Mientras que el MAE y el RMSE miden la magnitud de los errores absolutos, el $R^2$ proporciona una medida de cunta variabilidad de la variable dependiente (la variable a predecir) es explicada por el modelo.
<br><br>
El $R^2$ vara entre 0 y 1, y puede ser interpretado como el porcentaje de la variabilidad de la variable dependiente que es explicada por el modelo en comparacin con la variabilidad total presente en los datos.
<br><br>
La frmula para calcular el $R^2$ es la siguiente:
<br><br>
$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$
<br><br>
Donde:
<br><br>
<ul align="justify">
<li>\( SS_{res} \) es la suma de los cuadrados de los residuos (diferencia entre los valores reales y las predicciones del modelo).</li><br>
<li> \( SS_{tot} \) es la suma total de los cuadrados de la diferencia entre los valores reales y la media de la variable dependiente.</li><br>
</ul>
<p align="justify">
El $R^2$ puede tomar los siguientes valores:
<br><br>
<ul align="justify">
<li> $R^2 = 1$: El modelo ajusta perfectamente los datos, explicando el $100$% de la variabilidad de la variable dependiente.</li><br>
<li> $R^2 = 0$: El modelo no ofrece ninguna mejora en comparacin con simplemente usar la media de la variable dependiente para hacer predicciones.</li><br>
<li> $R^2 < 0$: El modelo se ajusta tan mal que su rendimiento es peor que simplemente usar la media como prediccin.</li><br>
</ul>
<p align="justify">
Es importante tener en cuenta que un alto $R^2$ no garantiza que el modelo sea vlido o que est libre de problemas.
<br><br>
El $R^2$ solo proporciona informacin sobre cmo el modelo se ajusta a los datos, pero no dice nada sobre la causalidad o la validez de las relaciones entre variables. Por lo tanto, es fundamental complementar la evaluacin con otras tcnicas y anlisis para asegurar la robustez y relevancia del modelo.

 ## **<font color="DeepPink">Residuos</font>**

<p align="justify">
En el contexto de modelos de regresin, los residuos son las diferencias entre los valores reales de la variable dependiente y las predicciones realizadas por el modelo. En otras palabras, los residuos son el error entre lo que el modelo predice y lo que realmente ocurre en los datos de prueba.
<br><br>
Matemticamente, el residuo de una observacin $i$ se calcula restando el valor real $y_{i}$ de la variable dependiente al valor predicho $\hat{y_{i}}$ por el modelo para esa misma observacin:
<br><br>
$$ \text{Residuo}_{i} = y_{i} - \hat{y_{i}}$$
<br>
<p align="justify">
Esperamos que los residuos tengan una distribucin alrededor de cero, lo que significa que el modelo ha ajustado bien los datos y que las predicciones estn lo ms cercanas posible a los valores reales.
<br><br>
Sin embargo, en la prctica, los residuos no siempre sern cero, ya que los modelos de regresin generalmente no pueden ajustarse perfectamente a todos los datos.
<br><br>
El anlisis de los residuos es una parte importante de la validacin del modelo. Algunas de las cosas que podemos hacer con los residuos son:
<br><br>
<ol align="justify">
<li><b>Comprobar la normalidad</b>: Un buen modelo de regresin debe tener residuos que sigan una distribucin normal. Si los residuos muestran una desviacin significativa de la normalidad, puede indicar que el modelo no est capturando completamente la estructura subyacente de los datos.</li>
<br>
<li><b>Identificar patrones no lineales</b>: Si los residuos muestran algn patrn sistemtico, como una forma de U o una relacin curvilnea con las variables independientes, esto puede indicar que el modelo no est capturando correctamente relaciones no lineales en los datos.</li>
<br>
<li><b>Verificar la homocedasticidad</b>: La homocedasticidad significa que los residuos tienen una varianza constante en todas las observaciones. Si los residuos tienen una varianza que cambia con el nivel de la variable dependiente, puede indicar una falta de homocedasticidad, lo que afecta la precisin y fiabilidad del modelo.</li>
<br>
<li><b>Detectar valores atpicos</b>: Los residuos tambin pueden ayudarnos a identificar valores atpicos o datos inusuales que pueden tener un impacto significativo en el modelo.</li></ol>
<p align="justify">
 En general, un anlisis detallado de los residuos es esencial para entender cmo bien se ajusta el modelo a los datos y si existen problemas o reas de mejora en el mismo.

 ## **<font color="DeepPink">Error</font>**

<p align="justify">
<b>Error</b> en el contexto de modelos y estadsticas, se refiere a la discrepancia entre un valor observado y el valor esperado o predicho por un modelo o estimacin. Es una medida de cunto el modelo o estimacin difiere de la realidad o los datos reales.
<br><br>
En un modelo de regresin, el error se refiere a la diferencia entre el valor real de la variable dependiente y el valor predicho por el modelo. Estos errores son lo que llamamos "residuos". En este contexto, el error indica cunto se desvan las predicciones del modelo de los valores reales en el conjunto de datos de prueba.
<br><br>
En otros contextos, como en el clculo de estimaciones o pronsticos, el error puede referirse a la diferencia entre el valor estimado o pronosticado y el valor real observado. En este caso, el error puede ser positivo (si la estimacin es mayor que el valor real) o negativo (si la estimacin es menor que el valor real).
<br><br>
El error es una parte esencial del proceso de modelado y estimacin, ya que nos permite evaluar la precisin y calidad de nuestros modelos o estimaciones. Un objetivo clave al construir modelos es minimizar el error y obtener predicciones o estimaciones lo ms precisas posible para los datos o situaciones en cuestin.

 ## **<font color="DeepPink">Importancia de Variables</font>**

<p align="justify">
La "Feature Importance" o importancia de variables o caracteristicas, es un concepto utilizado en el contexto del aprendizaje automtico (machine learning) y, en particular, en modelos de clasificacin o regresin. Se refiere a una tcnica que permite determinar la contribucin relativa de cada caracterstica o variable utilizada en el modelo para realizar las predicciones.
<br><br>
Cuando se entrena un modelo de aprendizaje automtico, como un algoritmo de clasificacin (por ejemplo, Random Forest, Gradient Boosting, etc.) o un algoritmo de regresin (por ejemplo, Regresin Lineal, Lasso, etc.), el modelo utiliza un conjunto de caractersticas o variables (tambin conocidas como atributos o columnas) para hacer predicciones.
<br><br>
La importancia de caractersticas nos ayuda a responder preguntas como:
<br><br>
<ul aling="justify">
<li>Cules son las caractersticas ms influyentes para el modelo en sus predicciones?</li><br>
<li>Cules caractersticas pueden ser descartadas sin afectar significativamente el rendimiento del modelo?</li><br>
<li>Cules caractersticas tienen una correlacin ms fuerte con la variable objetivo (o la variable a predecir)?</li></ul><br>
<p align="justify">
Hay varias formas de calcular la importancia de caractersticas, y algunos algoritmos ya proporcionan esta informacin directamente despus de ser entrenados.
<br><br>
Algunas de las tcnicas comunes para calcular la importancia de caractersticas son:
<br>
<ol align="justify">
<li><b>Importancia basada en rboles (Tree-based Importance)</b>:
   En algoritmos de rboles, como Random Forest o Gradient Boosting, la importancia de caractersticas se puede calcular al observar cmo cada caracterstica afecta la precisin de la prediccin en el rbol. Las caractersticas que se utilizan para tomar decisiones en los niveles superiores del rbol o <mark>que reducen la impureza</mark> (como la entropa o la ganancia de informacin) en mayor medida, generalmente se consideran ms importantes.</li><br>
<li><b>Coeficientes en regresin lineal</b>:
   En modelos de regresin lineal o regresin lineal regularizada (por ejemplo, Lasso, Ridge), los coeficientes asociados con cada caracterstica indican su contribucin a la prediccin. Los coeficientes ms grandes, en valor absoluto, indican que esa caracterstica tiene una mayor influencia en las predicciones.</li><br>
<li><b>Permutacin de caractersticas (Permutation Feature Importance)</b>:
   En este enfoque, se mide cmo cambia el rendimiento del modelo cuando se permutan aleatoriamente los valores de una caracterstica especfica mientras se evala en un conjunto de prueba. <mark>Si una caracterstica es importante, su permutacin causar una disminucin significativa en el rendimiento del modelo</mark>.</li></ol><br>
<p align="justify">
La importancia de caractersticas es una herramienta til para comprender y analizar el comportamiento de un modelo de aprendizaje automtico, as como para identificar qu caractersticas tienen un mayor impacto en las predicciones.
<br><br>
Sin embargo, debe utilizarse con precaucin, ya que la interpretacin de la importancia puede variar segn el contexto y el problema especfico que se est abordando.

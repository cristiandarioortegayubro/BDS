 ## **<font color="DeepPink">Para graficos</font>**

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

 # **<font color="DeepPink">Obtenci칩n de datos y generaci칩n del DataFrame</font>**

```python
data = "https://atm.mendoza.gov.ar/portalatm/zoneBottom/datosInteres/recaudacion/recaudacion_impuesto_ingresos.jsp"
lista = pd.read_html(data)
df = lista[0]
recaudacion = df.iloc[2:,1:24]
recaudacion.rename(columns = lambda col: f"A침o fiscal {col + 1999}", inplace=True)
recaudacion.index = recaudacion.index - 1
recaudacion.index.name = "Mes"
recaudacion = recaudacion.astype("Float32")
recaudacion 
```

# **<font color="DeepPink">Gr치ficos</font>**


## **<font color="DeepPink">Graficos para los primeros y 칰ltimos cinco ejercicios</font>**


```python
ejercicios = recaudacion.iloc[:, [0, 1, 2, 3, 4, 18, 19, 20, 21, 22]]
```

```python
fig = make_subplots(rows=2, cols=5, 
                    subplot_titles=[f'A침o {col[-4:]}' for col in ejercicios.columns],
                    x_title='Mes',
                    y_title='Recaudaci칩n')
```

```python
for i, (col_name, ej) in enumerate(ejercicios.items()):
  fig.add_trace(go.Scatter(x=ejercicios.index, 
                           y=ej, 
                           showlegend=True),
                row=i // 5 + 1, col=i % 5 + 1)
```

```python
fig.update_layout(height=660, width=1000, 
                  title_text="Ejercicios 2000-2004 y 2018-2022", 
                  showlegend=False,
                  template="gridon")

fig.show()
```

## **<font color="DeepPink">Evoluci칩n interanual</font>**


```python
df_annual_sum = ejercicios.sum().reset_index() \
                .rename(columns={'index': 'A침o', 0: 'Recaudaci칩n'}) \
                .replace('A침o fiscal', '', regex=True)
```

```python
fig = make_subplots(rows=1, 
                    cols=2, 
                    subplot_titles=['Per칤odo 2000-2004', 'Per칤odo 2018-2022'],
                    x_title='A침o',
                    y_title='Recaudaci칩n total')
```

```python
for i, df in enumerate([df_annual_sum.iloc[:5], df_annual_sum.iloc[-5:]]):
  fig.add_trace(go.Bar(x=df['A침o'], 
                       y=df['Recaudaci칩n'], 
                       showlegend=True),
  row=i // 2 + 1, col=i % 2 + 1)
```

```python
fig.update_layout( 
                  title_text="Recaudaci칩n total por a침o 2000-2004 y 2018-2022", 
                  showlegend=False,
                  template="gridon")

fig.show()
```

## **<font color="DeepPink">Comparativa del promedio de recaudaci칩n por mes, en cada per칤odo</font>**


```python
df_monthly_avg1 = ejercicios.T.iloc[:5].mean().reset_index() \
                  .rename(columns={0: 'Recaudaci칩n media'})
```

```python
df_monthly_avg2 = ejercicios.T.iloc[-5:].mean().reset_index() \
                  .rename(columns={0: 'Recaudaci칩n media'})
```

```python
fig = make_subplots(rows=1, 
                    cols=2, 
                    subplot_titles=['Per칤odo 2000-2004', 'Per칤odo 2018-2022'],
                    x_title='Mes de a침o',
                    y_title='Recaudaci칩n promedio')
```

```python
for i, df in enumerate([df_monthly_avg1, df_monthly_avg2]):
  fig.add_trace(go.Bar(x=df['Mes'], 
                       y=df['Recaudaci칩n media'], 
                       showlegend=True),
  row=i // 2 + 1, col=i % 2 + 1)
```

```python
fig.update_layout(title_text="Recaudaci칩n media de cada mes del a침o, en los per칤odos 2000-2004 y 2018-2022", 
                  showlegend=False,
                  template="gridon")

fig.show()
```
